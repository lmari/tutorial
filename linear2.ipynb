```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_coppie = 100
X = np.random.normal(170, 8, num_coppie)
Y = (X - 100) + np.random.normal(0, 8, num_coppie)

plt.figure(figsize=(8, 6))
plt.plot(X, Y, 'o')
plt.xlabel('altezza [cm]', fontsize=14)
plt.ylabel('peso [kg]', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()
```

```python
x_medio = np.mean(X)
y_medio = np.mean(Y)
k1 = (np.sum((X - x_medio) * (Y - y_medio))) / np.sum((X - x_medio) ** 2)
k0 = y_medio - k1 * x_medio
mse = np.mean(((k0 + X * k1) - Y) ** 2)

plt.figure(figsize=(8, 6))
plt.plot(X, Y, 'o')
plt.plot(X, k0 + k1 * X)
plt.xlabel('altezza [cm]', fontsize=14)
plt.ylabel('peso [kg]', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title(f'k0 = {k0:.2f}, k1 = {k1:.2f}, MSE = {mse:.1f}')
plt.grid(True)
plt.show()
```
